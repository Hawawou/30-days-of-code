```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
```

A simple starter plot

```python
x_data = [1,2,3,4,5,6,7,8,9]
y_data = np.exp(x_data)
```

```python
plt.plot(x_data, y_data)
plt.title('Starter plot')
plt.xlabel('X coordinates')
plt.ylabel('Y coordinates')
plt.figure(figsize=(10,8))
```

Subplots
plt.subplot displays multiple plots in thesame figure
plt.subplot(num of rows, number of columns, plot number)

```python
plt.subplot(1,3,1)
plt.plot(x_data, y_data)

plt.subplot(1,3,2)
plt.plot(x_data, np.sin(y_data))

plt.subplot(1,3,3)
plt.plot(x_data, np.cos(y_data))

# plt.subplots(figsize=(10,5))
```

```python
x = np.linspace(0, 50, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1)
plt.title('this is y1')
plt.xlabel('x coordinates')
plt.ylabel('y coordiantes')
```

```python
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(loc='lower left')
```

Lines color and style

```python
plt.figure(figsize=(10,8))

plt.plot(x, y1, color='red')
plt.plot(x, y2, color='y')
plt.plot(x, y2*2, color='#00FF00')
plt.plot(x, y2*3, color='0.8')
```

```python
# Line styles
plt.figure(figsize=(10,8))

plt.plot(x, y1, color='red', ls='solid', marker='+', lw=3, markersize=3)
plt.plot(x, y2, color='b', ls='dotted', marker='D', lw=4, markersize=4)
plt.plot(x, y2*2, color='0.8', ls=':', marker='s', lw=5, markersize=5)
```

Text and annotation

```python
x = [1,2,3,4,5,6,7,8]
y = np.square(x)

plt.plot(x,y)

plt.text(2, 60, 'This is added at the coordinate (2,60)', color='red')
```

```python
# Text annotation
plt.plot(x,y)

plt.annotate('This is the minumum point', 
              xy=(1,1), 
              xytext=(1.4, 30), 
              arrowprops=dict(facecolor='red'))

plt.annotate('This is the maximum point', 
              xy=(8,64), 
              xytext=(1,55), 
              bbox=dict(boxstyle='round'),
              arrowprops=dict(arrowstyle='fancy', connectionstyle='angle3, angleA=0, angleB=-90'))

plt.annotate('Middle, estimated',
              xy=(4.5,22),
              xytext=(5,12),
              bbox=dict(boxstyle='round', fc='none'),
              arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3))
```

Object oriented method

```python
# Create an empty figure
fig = plt.figure()

# create axes
ax = plt.axes()

# Plot on that axes
ax.plot(x, y, 'g' )

ax.set_title('Object oriented')

ax.set_xlabel('x coordinates')
ax.set_ylabel('y coordinates')
```

```python
#  for subplots

# create axes
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot
axes[0].plot(x, y, 'g')
axes[1].plot(x, y, 'b')

axes[0].set_title('Title 1')
axes[1].set_title('Title 2')

```

```python
# iterating through axes

fig, axes = plt.subplots(nrows=1, ncols=3)
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('X label')
    ax.set_ylabel('Y label')
    ax.set_title('Title')

    fig

# To avoid overlap of the elements of figures
plt.tight_layout()
```

```python
ax = plt.axes()

ax.plot(x, y, 'g')
ax.set(title='Title',
       xlabel='x label',
       ylabel='ylabel',
       ylim=(-10, 10),
       xlim=(0,25)); #end with ; to get rid of unused figure details
```

Types of plots

```python
# Scatter plot: used to show th relationship btn 2 variables

plt.scatter(y1, y2, marker='s')
```

```python
plt.scatter(y, x, marker='>')
```

```python
# Bar plot: display relationship btn 2 variables(one numerical and one categorical)

qty = np.array([30,30,30,10])
fruits = ['apple', 'orange', 'lemon', 'pineapple']

plt.bar(fruits, qty)
```

```python
# Pie charts: show proportion of a feature element
plt.pie(qty, labels=fruits, shadow=True)
```

```python
# Histograms: show frq distribution of data values
fruits = ['apple', 'orange', 'lemon', 'pineapple', 'orange', 'pineapple']
plt.hist(fruits);
```

```python
data = np.random.randn(100)
plt.hist(data);
```
