```python
# while loops
n = 5
while n > 0:
  n -= 1
  print(n)
```

```python
# break and continue 
n = 5
while n > 0:
  n -= 1
  if n == 2:
    break
  print(n)
print("Loop ended!")
```

```python
n = 5 
while n > 0:
  n -= 1
  if n == 2:
    continue
  print(n)
print("Loop ended!")
```

```python
# else clause
n = 5
while n > 0:
  n -= 1
  print(n)
else:
  print('Loop ended')
```

```python
n = 5
while n > 0:
  n -= 1
  print(n)
  if n == 2:
    break
else: 
  print('Loop done')
```

```python
# nested while loops
a = ['foo', 'bar']
while len(a):
  print(a.pop(0))
  b = ['baz', 'qux']
  while len(b):
    print('>', b.pop(0))
```
